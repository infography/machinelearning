```python
featto_use=[]
for col in con1_df.columns:
    if col not in ob_col:
        featto_use.append(col)
```

```python
featto_use.remove('logerror')
```

```python
con2_df = con1_df[featto_use]
```

```python
X_train,y_train=con2_df,con1_df['logerror']
```

```python
X_train.shape
```

```python
y_train.dtype
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
clf = GradientBoostingRegressor()
```

```python
clf = clf.fit(X_train,y_train)
```

```python
clf.feature_importances_[1]
```

```python
feat_imp_num = list(clf.feature_importances_)
```

```python
feat_dict = dict(zip(featto_use,feat_imp_num))
```

```python

sorted_feat = reversed(sorted(feat_dict,key=feat_dict.__getitem__))
```

```python
feat_name_rank=[]
feat_rank=[]
for k in sorted_feat:
    if feat_dict[k]!=0.0:
        feat_name_rank.append(k)
        feat_rank.append(feat_dict[k])
    
```

```python
rankwise_dict = dict(zip(feat_name_rank,feat_rank))
```

```python
feat_df = pd.DataFrame(rankwise_dict.items())
# pd.set_option.max_columns=400
```

```python
feat_df.columns = ['f_name','rank']
```

```python
sns.set(font_scale=1.5)
plt.figure(figsize=(10,12))
```

```python
sns.barplot(x='rank',y='f_name',data=feat_df)
```

```python
sns.plt.show()
```

```python
cor_list=[]
for col in feat_name_rank:
    cor_list.append(np.corrcoef(con1_df[col],con1_df['logerror'])[0][1])
    
```

```python
cor_dict = dict(zip(feat_name_rank,cor_list))
```

```python
cor_df = pd.DataFrame(cor_dict.items())
```

```python
cor_df.columns = ['col_name','corr_val_with_logerror']
```
